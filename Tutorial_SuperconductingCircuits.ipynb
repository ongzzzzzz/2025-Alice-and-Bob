# ABC of Hamiltonian engineering with superconducting Josephson junctions

Let's explore a bit how we can use Josephson junctions as non-linear elements in a superconducting circuit to construct Hamiltonians that we want.

## Building blocks of quantum circuits

### The variables $Q$ and $\Phi$

As in the classical network analysis, we can use current ($I$) and voltage ($V$) to describe these circuit elements. However, we are going to use charge ($Q$) and flux ($\Phi$) variables which are more convenient because they are the natural variables when you quantize a quantum system.

They are related to the current and voltage through Coulomb's law and Weber's law:

$$
\begin{align}
Q(t)  &= \int_{-\infty}^{t}I(t')dt' \quad (\textrm{Coulomb}) \\ 
\Phi &= \int_{-\infty}^{t}V(t')dt' \quad (\textrm{Weber})
\end{align}
$$

Furthermore, we are going to use the reduced flux and charge variables: 

$$\varphi =2 \pi \frac{\Phi}{\Phi_0} \quad \quad \quad n =\frac{Q}{2e}$$

where $\phi_0 = \frac{h}{2e} \approx 2.067 \times 10^{-15}$ Wb is the magnetic flux quantum and $2e$ corresponds to a pair of electrons ($2e$) called Cooper pair which carries the current in superconductors.

### Linear elements
The two basic linear building blocks in classical as well as superconducting circuits are capacitor (C) and inductor (L).

Their potential energies are :

- Capacitor:
 $$ E = 4 E_Cn^2  \quad \quad \quad  E_C = \frac{e^2}{2C}$$

- Inductor:
$$ E = E_L \frac{\varphi ^2}{2}  \quad \quad \quad  E_L = \frac{(\Phi_0/2 \pi)^2}{L} $$

### The Josephson junction

When two superconductors are brought in close proximity it turns out that Cooper pairs can tunnel through the weak link and form a supercurrent. This effect was ignored until 1960s as it was thought that such event would require simultaneous tunneling of two incoherent electrons which would make the observation unlikely. In 1962 Brian Josephson predicted that Cooper pair tunneling was a coherent event and the resultant supercurrent was in the same order of magnitude as quasiparticle currents. Josephson’s prediction was followed by the experimental proof in 1963 and the Josephson junction became the generic name for weakly coupled superconductors. A plethora of device configurations were shown to display Josephson effect. Here are three examples:

1. Superconductor (S) - Insulator (I) - Superconductor (S)
2. Superconductor (S) - Normal metal (N) - Supercondutor (S)
3. Superconductor - Graphene - Superconductor

![](media/JJs.png "Different types of Josephson junctions")


Two very important relations to know about Josephson Junctions are :

**DC Josephson effect**

$$
I_s(\varphi) = I_c \sin(\varphi)
$$

where $I_c$ is the critical supercurrent that the junction can withstand and $\varphi$ is the phase across the junction.

**AC Josephson effect**

$$
\dot{\varphi} = \frac{2 \pi}{\Phi_0}V
$$

AC/DC Josephson effects are the fundamental relations of a Josephson junction as they relate current & amp; phase and voltage. Here we are going to mostly deal with the energy stored in a Josephson junction. Note that it is not linear, which make Josephson Junction very useful, as we will see in the following lectures.

**Josephson energy**
$$ E(\varphi) = -E_{J}\cos(\varphi)$$

where $E_J= \frac{I_c \Phi_0}{2 \pi}$ is the Josephson energy. The potential landscape of a Josephson junction is plotted in the figure below (black). Note that for small displacement in $\varphi$, the potential can be approximated by that of a harmonic oscillator (dashed red). Taylor expanding the potential at $\varphi = 0$

$$
E = -E_J (1-\frac{\varphi^2}{2!}+\frac{\varphi^4}{4!}+ h.o)
$$

$$
E = \underbrace{-E_J}_{\text{constant offset}} +\underbrace{E_j\frac{\varphi^2}{2}}_{\text{linear inductor}}-E_j\frac{\varphi^4}{24}+ h.o
$$

![](media/JJ_Potential.png "Josephson junction potential")

we see that the linear term represents a Josephson inductance $L_J = \frac{\Phi_0}{2\pi I_c} = \frac{(I_0/ 2 \pi)^2}{E_J}$

## Method to quantize and analyze a circuit

Let us illustrate how we can quantize a given superconducting circuit using the following example:

![](media/example_circuit.png "Capacitor and Josephson junction in a closed circuit")

The energies at play are (using the formula for the two independent components, as expressed earlier in this lecture) :
$$
E = 4 E_Cn^2 \quad \quad \quad E(\varphi) = -E_{j}\cos(\varphi)
$$

We promote the reduced flux and charge variables to quantum operators:

$$
(\varphi,n) \quad \rightarrow \quad (\hat{\varphi}, \hat{n})
$$

So the Hamiltonian reads :
$$
H = 4E_C\hat{n}^2-E_j\cos\hat{\varphi}
$$

Note that the variables form a conjugate pair 
$$
[\varphi,n] = i\hbar
$$

Initially, this circuit was operated in the regime $(E_j \sim E_c)$ and called a Cooper pair box. The number of charges on the island is then a well defined number and exploited for information encoding. Over the years, due to charge noise issues, the community converged towards the regime $(E_j \gg E_c)$ where the same configuration is more robust against the charge noise. In this regime, the device is called as transmon and is the most commonly used qubit configuration nowadays.


### Hamiltonian diagonalization 

We can rewrite the Hamiltonian separating the linear and non-linear parts as

$$
\begin{align*}
H & = H_{\textrm{linear}} + H_{\textrm{non-linear}} \\
H & = \underbrace{(4E_C\hat{n}^2 + \frac{E_j}{2}\varphi^2)}_{\text{linear}} -\underbrace{E_j(\cos\varphi + \frac{\varphi^2}{2})}_{\text{non-linear}}
\end{align*}
$$

on the linear part, we can jump to the second quantization and define the conjugate variables in terms of annihilation &amp; creation operators as:

$$

\hat{n} = n_{zpf} \times i(\hat{a}-\hat{a}^{\dagger}) \quad \quad \hat{\varphi} = \varphi_{zpf} \times (\hat{a}+\hat{a}^{\dagger})

$$

where $n_{zpf} =[E_L/(32E_C)]^{1/4}$ and $\varphi_{zpf}=(2E_C/E_L)^{1/4}$ are the zero-point fluctuations. Heisenberg uncertainty principle sets a lower limit on the product of fluctuations of conjugate variables. In terms of coupling of modes, the participation of a mode in the non-linear element is measured by its contribution on the zero point fluctuations of the relevant variable.

Plugging the newly defined operators into (17) and re-arranging, we obtain

$$

H= \underbrace{\hbar \omega_a a^{\dagger}a}_{\text{linear }} - \frac{E_J}{4!}\varphi_{zpf}^4(\hat{a}+\hat{a}^{\dagger})^4 + h.o

$$

### With several modes participating in nonlinearity

For now, we have only consider one mode. However, several modes can coexist and participate to a non-linearity. To find the Hamiltonian form of such a system, we adopt a blackbox approach. Regardless of the circuit topology, we consider the system in its diagonalized form and we only care about the two aspects of the independents linear modes:

1. frequency of each linear mode: $\omega_i$
2. participation in the zero-point fluctuations of the non-linearity (here a Josephson Junction) : $\varphi_i$ 

Then, the full Hamiltonian takes the form:

$$
\begin{align*}
H &= \sum_{i}\hbar \omega_i a_i^{\dagger}a_i -  E_{J}\sum_i[(\cos(\varphi_i(a_i +a^{\dagger}_i)) + \frac{1}{2}(\varphi_i(a_i +a^{\dagger}_i))^2]\\
& = \sum_{i}\hbar \omega_i a_i^{\dagger}a_i -  E_{J}\sum_i \cos_4(\varphi_i(a_i +a^{\dagger}_i))
\end{align*}
$$

This black box approach is useful, because in practise, we make use of finite element EM simulations to obtain the mode frequencies and their participation in zpf.

Here is a schematic of the method we use, where we get the non-linear part by getting the diagonalize form of the circuit with one mode. We then consider the full multi-mode system.

### General recipe

**Nonlinear element zoo**: We have seen the example of a single Josephson junction (JJ) for the non-linear element. It has a cosine potential which yields only the even powers of nonlinearity. We can come up with more complex nonlinear elements by combining several JJs. Here is a zoo of possibilities:

| Name    | Element | Potential Energy |
| -------- | ------ | -------- |
| Josephon Junction | ![](media/JJ.png) | $-E_J \cos(\varphi)$|
| RF-SQUID | ![](media/RFSQUID.png) | $-E_J \cos(\varphi) + \frac{E_L}{2} (\varphi - \varphi_\mathrm{ext})^2$ |
| DC-SQID | ![](media/DCSQUID.png) | $-E_{J_1} \cos(\varphi) - E_{J_2}\cos(\varphi - \varphi_\mathrm{ext})$|
| JJ chain | ![](media/ChainJJ.png) | $-NE_J \cos(\varphi/N)$ (note the role of N for the higher order terms. It’s going to reduce the nonlinearity !) |
| SNAIL (Superconducting Nonlinear Asymmetric Inductance eLement) | ![](media/SNAIL.png) | $-\alpha E_J \cos(\varphi - \varphi_\mathrm{ext}) - N E_J \cos(\varphi/N)$ |
| ATS (Asymmetrically Threaded SQUID) | ![](media/ATS.png) | $- 2 E_J \cos(\varphi_\Sigma)\cos(\varphi + \varphi_\Delta) - E_L\varphi^2/2$ |
